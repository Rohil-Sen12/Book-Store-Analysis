```python
import plotly.express as px

top_authors = toolkit.groupby("Author", as_index=False)["Stock"].sum()
top_authors = top_authors.sort_values(by="Stock", ascending=False).head(5)

fig = px.bar(top_authors, x="Author", y="Stock", color="Author",
             title="Top 10 Authors by Stock",template="plotly_dark")
fig.show()
```

```python
fig = px.pie(toolkit, names="Genre", values="Stock", title="Stock Distribution by Genre",color="Stock",template="plotly_white")
fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.barplot(data=toolkit, x="Genre", y="Stock", estimator=sum, ci=None,palette="viridis")
plt.xticks(rotation=45)
plt.title("Total Stock by Genre")
plt.show()
```

```python
yearly_stock = toolkit.groupby("Published_Year", as_index=False)["Stock"].sum()
fig = px.line(yearly_stock, x="Published_Year", y="Stock",
              title="Stock Trend by Published Year",template="plotly_dark")
fig.show()
```
